```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("eval.xlsx", sheet_name=None)
keys = list(df.keys())

for i, key in enumerate(keys):
    acc = df[key]["acc"]
    bi_acc = df[key]["bi_acc"]
    plt.scatter(np.ones_like(acc)*i, acc, alpha=0.5)
    plt.scatter(np.ones_like(acc)*i, bi_acc, alpha=0.5)

keys
```

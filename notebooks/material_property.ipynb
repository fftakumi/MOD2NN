```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy
```

## Bi2.5NIGG

```python
faraday = pd.read_excel("data/Ms   bi2.5nigg 10 layers glass sub..xlsx")
ellipticity = pd.read_excel("data/PM  Ms  bi2.5nigg 10 layers glass sub..xlsx")
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(131)
ax1.plot(faraday["Wavelength"], faraday["R"])
ax2 = fig.add_subplot(132)
ax2.plot(ellipticity["Wavelength"], -np.tan(ellipticity["E"] * np.pi/180))
ax3 = fig.add_subplot(133)
ax3.plot(faraday["Wavelength"], faraday["R"])
ax3.plot(ellipticity["Wavelength"], -ellipticity["E"])
ax2.set_ylim(-1, 1)
```

```python
from scipy.interpolate import interp1d

wavelength = 633
fit_faraday = interp1d(faraday["Wavelength"], faraday["R"])
fit_ellipticity = interp1d(ellipticity["Wavelength"], -ellipticity["E"])
print("faraday:", fit_faraday(wavelength), "ellipticity:", fit_ellipticity(wavelength))
print(np.tan(fit_ellipticity(wavelength) * np.pi / 180))
```

```python
x = np.linspace(400, 1000,1000)
plt.plot(np.tan(fit_ellipticity(x) * np.pi / 180), fit_faraday(x) * np.pi/180)
```

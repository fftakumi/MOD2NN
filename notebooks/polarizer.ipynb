```python
import numpy as np
import matplotlib.pyplot as plt
```

$$
\begin{aligned}
  R(\phi)T_{rcp}R(-\phi)&=
  \begin{pmatrix}
    \cos(\phi) & -\sin(\phi)\\
    \sin(\phi) & \cos(\phi)
  \end{pmatrix}
  \dfrac{1}{2}
  \begin{pmatrix}
    1 & 0\\
    i & 0
  \end{pmatrix}
  \begin{pmatrix}
    \cos(\phi) & \sin(\phi)\\
    -\sin(\phi) & \cos(\phi)
  \end{pmatrix}\\

  &=
  \dfrac{1}{2}
  \begin{pmatrix}
    \cos(\phi) & -\sin(\phi)\\
    \sin(\phi) & \cos(\phi)
  \end{pmatrix}
  \begin{pmatrix}
    \cos(\phi) & \sin(\phi)\\
    i\cos(\phi) & i\sin(\phi)
  \end{pmatrix}\\

  &=\dfrac{1}{2}
  \begin{pmatrix}
    \cos^2(\phi)-i\sin(\phi)\cos(\phi) & \cos(\phi)\sin(\phi)-i\sin^2(\phi)\\
    \sin(\phi)\cos(\phi)+i\cos^2(\phi) & \sin^2(\phi)+i\cos(\phi)\sin(\phi)
  \end{pmatrix}\\

  &=\dfrac{1}{2}
  \begin{pmatrix}
    \cos^2(\phi) & \cos(\phi)\sin(\phi)\\
    \sin(\phi)\cos(\phi) & \sin^2(\phi)
  \end{pmatrix}
  +
  \dfrac{1}{2}
  \begin{pmatrix}
    -i\sin(\phi)\cos(\phi) & -i\sin^2(\phi)\\
    i\cos^2(\phi) & i\cos(\phi)\sin(\phi)
  \end{pmatrix}\\

  &=\dfrac{1}{2}
  \begin{pmatrix}
    \cos^2(\phi) & \dfrac{1}{2}\sin(\phi)\\
    \dfrac{1}{2}\sin(\phi) & \sin^2(\phi)
  \end{pmatrix}
  +
  \dfrac{i}{2}
  \begin{pmatrix}
    -\dfrac{1}{2}\sin(\phi) & -\sin^2(\phi)\\
    \cos^2(\phi) &  \dfrac{1}{2}\sin(\phi)
  \end{pmatrix}\\
\end{aligned}
$$

```python
phi = np.pi/2
phi_pol = np.linspace(0, 2*np.pi, 100)
T = np.array([[np.cos(phi_pol)**2, np.sin(2*phi_pol)/2],
              [np.sin(2*phi_pol)/2, np.sin(phi_pol)**2]])

r_hat = np.array([1,1.j]).reshape((-1, 1))
l_hat = np.array([1,-1.j]).reshape((-1, 1))

rcp = 2.0*r_hat * np.exp(-1.j*phi) / np.sqrt(2)
lcp = l_hat * np.exp(1.j*phi) / np.sqrt(2)

Ein = rcp + lcp

Eout = []
for i in range(len(phi_pol)):
    Eout.append(T[:,:,i] @ Ein)

Eout = np.array(Eout)
Iout = np.abs(Eout[:,0,0])**2/2 + np.abs(Eout[:,1,0]**2/2)
plt.plot(phi_pol, Eout[:,1,0].real)
plt.plot(phi_pol, Eout[:,1,0].imag)
plt.xticks(np.arange(0, 2*np.pi+0.1, np.pi/2), ["0", r"$\pi/2$", r"$\pi$", r"$3\pi/2$", r"$2\pi$"])
```

```python
phi=np.pi/2
T = np.array([[1,0],
             [0,0]])

phi_pol = np.linspace(0, 2*np.pi, 100)
R=np.array([[np.cos(phi_pol), -np.sin(phi_pol)],
           [np.sin(phi_pol), np.cos(phi_pol)]])

r_hat = np.array([1,1.j]).reshape((-1, 1))
l_hat = np.array([1,-1.j]).reshape((-1, 1))

rcp = 2.0*r_hat * np.exp(-1.j*phi) / np.sqrt(2)
lcp = l_hat * np.exp(1.j*phi) / np.sqrt(2)

Ein = rcp + lcp
Tr = np.array([[1/2,0],
             [1.0j/2,0]])
Tl = np.array([[1/2,0],
             [-1.0j/2,0]])
Er_out = []
El_out = []
Eout = []
for i in range(len(phi_pol)):
    Er_out.append(R[:,:,i] @ Tr @ R[:,:,i].T @ Ein)
    El_out.append(R[:,:,i] @ Tl @ R[:,:,i].T @ Ein)

Er_out = np.array(Er_out)
El_out = np.array(El_out)
Eout = np.array(Er_out + El_out)
Iout = np.abs(Eout[:,0,0])**2/2 + np.abs(Eout[:,1,0]**2/2)
plt.plot(phi_pol, Eout[:,1,0].real)
plt.plot(phi_pol, Eout[:,1,0].imag)
plt.xticks(np.arange(0, 2*np.pi+0.1, np.pi/2), ["0", r"$\pi/2$", r"$\pi$", r"$3\pi/2$", r"$2\pi$"])
```

```python
phi=np.pi/2

phi_pol = np.linspace(0, 2*np.pi, 100)
Tr = 1/2*np.array([[np.cos(phi_pol)**2, np.sin(2*phi_pol)/2],
              [np.sin(2*phi_pol)/2, np.sin(phi_pol)**2]]) + \
    1.0j/2*np.array([[-np.sin(2*phi_pol)/2, -np.sin(phi_pol)**2],
                  [np.cos(phi_pol)**2, np.sin(2*phi_pol)/2]])

Tl = 1/2*np.array([[np.cos(phi_pol)**2, np.sin(2*phi_pol)/2],
              [np.sin(2*phi_pol)/2, np.sin(phi_pol)**2]]) + \
    -1.0j/2*np.array([[-np.sin(2*phi_pol)/2, -np.sin(phi_pol)**2],
                  [np.cos(phi_pol)**2, np.sin(2*phi_pol)/2]])

r_hat = np.array([1,1.j]).reshape((-1, 1))
l_hat = np.array([1,-1.j]).reshape((-1, 1))

rcp = 2.0*r_hat * np.exp(-1.j*phi) / np.sqrt(2)
lcp = l_hat * np.exp(1.j*phi) / np.sqrt(2)

Ein = rcp + lcp
Er_out = []
El_out = []
Eout = []
for i in range(len(phi_pol)):
    Er_out.append(Tr[:,:,i] @ Ein)
    El_out.append(Tl[:,:,i] @ Ein)

Er_out = np.array(Er_out)
El_out = np.array(El_out)
Eout = np.array(Er_out + El_out)
Iout = np.abs(Eout[:,0,0])**2/2 + np.abs(Eout[:,1,0]**2/2)
plt.plot(phi_pol,Iout, label="I")
plt.plot(phi_pol, Er_out[:,0,0].real, label="real")
plt.plot(phi_pol, Er_out[:,0,0].imag, label="imag")
plt.legend()
plt.xticks(np.arange(0, 2*np.pi+0.1, np.pi/2), ["0", r"$\pi/2$", r"$\pi$", r"$3\pi/2$", r"$2\pi$"])
```
